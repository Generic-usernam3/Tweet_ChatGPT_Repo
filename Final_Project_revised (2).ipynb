Functions for removing punctuation, "\n" , hyperlinks, hashtags etc.

```python

def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text
def remove_the_n(text):
    text = re.sub('\n', '', text)
    return text 
def remove_hyper(text):
    text = re.sub(r'http\S+', '',text)
    return text 
def remove_hash(text):
    text = re.sub("#([a-zA-Z0-9_]{1,50})", '', text)           
    return text
def remove_nonunicode(text):
    text = re.sub
stop = text.ENGLISH_STOP_WORDS        
```

```python
map_dict = dict()          #switch string label to 0 or 1 

map_dict['bot']=1
map_dict['Not a bot']=0
```

```python
mixed_tweets['Text'] = mixed_tweets['Text'].apply( lambda x: remove_the_n(x))                                              #remove "\n"
mixed_tweets['Text'] = mixed_tweets['Text'].apply( lambda x: remove_hyper(x))                                              #remove hyperlink
mixed_tweets['Text'] = mixed_tweets['Text'].apply( lambda x: remove_hash(x))                                               #remove hashtag
mixed_tweets['Text'] = mixed_tweets['Text'].apply(lambda x: remove_punct(x))                                               #remove punctuation
mixed_tweets = mixed_tweets.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))               #remove emoji
#mixed_tweets['Text'] = mixed_tweets['Text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))  #remove stopwords
mixed_tweets.head(20)
mixed_tweets = mixed_tweets.applymap(lambda s: s.encode('utf-8').decode('utf-8') if isinstance(s, str) else s)             #remove hidden non-unicode characters
mixed_tweets=mixed_tweets[mixed_tweets['Text'].str.strip() != ''][mixed_tweets['Text'].notnull()]                          #drop NAN + empty cells
mixed_tweets['Type'] = mixed_tweets['Type'].map(map_dict).astype('int')                                                    #string number to number
mixed_tweets['Text'] = mixed_tweets['Text'].apply(str.lower)                                                               #to lowercase
```

```python
mixed_tweets
```

Splitting the tweets into training and test dataframes

```python
finetune, test = train_test_split(mixed_tweets, test_size=0.2)

finetune = finetune[['Text', 'Type']]

test = test[['Text', 'Type']]
```

```python
finetune = finetune.reset_index(drop=True)
test = test.reset_index(drop=True)
```

💢  switching the column labels to be compatible with BERT  💢
          This was a pain in the *$% to figure out

```python
finetune.rename(columns={'Text': 'text', 'Type': 'label'}, inplace=True)
test.rename(columns = {'Text': 'text', 'Type': 'label'}, inplace=True)
```

```python
  

dataset_dict = {'finetune': Dataset.from_pandas(finetune),
                "test": Dataset.from_pandas(test)}
dataset = DatasetDict(dataset_dict)
```

```python
dataset
```

##tokenizing tweets

```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')
```

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], 
                     padding="max_length", 
                     truncation=True, 
                     max_length=500)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
tokenized_datasets
```

```python
train_dataset = tokenized_datasets["finetune"]
eval_dataset = tokenized_datasets["test"]
```

##model building

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained('bert-base-cased', 
                                                           num_labels=2)
```

```python

from transformers import TrainingArguments

training_args = TrainingArguments(num_train_epochs=1,
                                  output_dir="output")
```

```python
import numpy as np
from datasets import load_metric

metric = load_metric("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    accuracy = metric.compute(predictions=predictions,
                             references=labels)
    return accuracy
```

```python
from transformers import Trainer

trainer = Trainer(
     model=model,
     args=training_args,
     train_dataset=train_dataset,
     eval_dataset=eval_dataset,
     compute_metrics=compute_metrics)
     
```

```python
trainer.train()
```

```python
trainer.evaluate()
```
